```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
box_office = pd.read_csv('asia_pacific_box_office.csv')
movie_info = pd.read_csv('movie_info.csv')
print(box_office.head())
print(movie_info.head())
```

```python
movie_info.loc[ movie_info['worldwide_collection_in_million_(USD)'] >= 1000, "suc" ] = 1
movie_info.loc[ movie_info['worldwide_collection_in_million_(USD)'] < 1000, "suc" ] = 0
```

```python
success = []
for num in movie_info['worldwide_collection_in_million_(USD)']:
    if num >= 1000: 
        success.append('1')
    else: 
        success.append('0')
        
movie_info['success'] = success
movie_info.head(10)
```

```python
box_office.shape, movie_info.shape
```

```python
movie_info.head()
```

```python
# South Korea 추가
movie_info['South_Korea'] = box_office['South Korea']
movie_info
```

```python
sel = ['South_Korea']

X = movie_info[sel]
y = movie_info['success']

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.2,
                                                   random_state = 0)

X_train.shape, X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
model1 = KNeighborsClassifier()
model1.fit(X_train, y_train)
print("학습용 정확도 : ", model1.score(X_train,y_train))
print("테스트용 정확도 : ", model1.score(X_test,y_test))

model2 = DecisionTreeClassifier(max_depth=1, random_state=0)
model2.fit(X_train, y_train)
print("학습용 정확도 : ", model2.score(X_train,y_train))
print("테스트용 정확도 : ", model2.score(X_test,y_test))

model3 = RandomForestClassifier(max_depth=1, n_estimators=500, random_state=0)
model3.fit(X_train, y_train)
print("학습용 정확도 : ", model3.score(X_train,y_train))
print("테스트용 정확도 : ", model3.score(X_test,y_test))
```

```python
# 평균평점 colomn 추가하기

avg_rating = []
for rate in range(len(movie_info)) :
    list = []
    list.append(movie_info['tomatometer'][rate])
    list.append(movie_info['tomato_audience_score'][rate])
    list.append(movie_info['imdb'][rate])
    list.append(movie_info['metascore'][rate])
    list.append(movie_info['meta_user_score'][rate])
    avg_rating.append(round(np.mean(list),1))
print(avg_rating)
movie_info['avg_rating'] = avg_rating
print(movie_info['avg_rating'])
```

```python
# 평균평점과 흥행수익의 상관관계

sel = ['avg_rating']

X = movie_info[sel]
y = movie_info['worldwide_collection_in_million_(USD)']

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.3,
                                                   random_state = 0)

X_train.shape, X_test.shape
```

```python
model1 = KNeighborsClassifier()
model1.fit(X_train, y_train)
print("학습용 정확도 : ", model1.score(X_train,y_train))
print("테스트용 정확도 : ", model1.score(X_test,y_test))

model2 = DecisionTreeClassifier(max_depth=1, random_state=0)
model2.fit(X_train, y_train)
print("학습용 정확도 : ", model2.score(X_train,y_train))
print("테스트용 정확도 : ", model2.score(X_test,y_test))

model3 = RandomForestClassifier(max_depth=1, n_estimators=500, random_state=0)
model3.fit(X_train, y_train)
print("학습용 정확도 : ", model3.score(X_train,y_train))
print("테스트용 정확도 : ", model3.score(X_test,y_test))
```
